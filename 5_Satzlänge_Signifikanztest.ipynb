# Durchschnittliche Satzlänge und Signifikanztest


## Bei männlichen Autoren

```python
import os
import re


corpus_directory = "/Users/a1/Documents/0. UNI/6. SS 2023/Forsch_design/projekt/neu/männlich_txt"

average_sentence_lengths_m = []

# um Sätze zu erkennen
sentence_pattern = r"[.!?]"


for filename in os.listdir(corpus_directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(corpus_directory, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()

        sentences = re.split(sentence_pattern, text)
        num_sentences = len(sentences)

        total_sentence_length = sum(len(sentence.split()) for sentence in sentences)
        average_length_m = total_sentence_length / num_sentences

        # Für Signifikanztest nötig!!!!
        average_sentence_lengths_m.append(average_length_m)

# print("Durchschnittliche Satzlängen in allen Dateien:")
# for avg_length in average_sentence_lengths_m:
#    print(avg_length)
```

## Bei weiblichen Autoren

```python
import os
import re

corpus_directory = "/Users/a1/Documents/0. UNI/6. SS 2023/Forsch_design/projekt/neu/weiblich_txt"

average_sentence_lengths_w = []
sentence_pattern = r"[.!?]"

for filename in os.listdir(corpus_directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(corpus_directory, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()


        sentences = re.split(sentence_pattern, text)
        num_sentences = len(sentences)


        total_sentence_length = sum(len(sentence.split()) for sentence in sentences)
        average_length_w = total_sentence_length / num_sentences

        average_sentence_lengths_w.append(average_length_w)
```

```python
print(average_length_w)
```

## Signifikanz Prüfen

```python
import scipy.stats as stats

t_statistic, p_value = stats.ttest_ind(average_sentence_lengths_w, average_sentence_lengths_m) #list

print("T-Statistik:", t_statistic)
print("P-Wert:", p_value)
```
