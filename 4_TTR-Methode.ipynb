# Komplexität der Sätze mit TTR ausrechnen

## Zusammenhang der Types und Tokens (TTR)

```python
import os
import nltk
from nltk.tokenize import word_tokenize

def calculate_ttr(text):
    tokens = word_tokenize(text)
    types = set(tokens)
    ttr = len(types) / len(tokens)
    return ttr

male_folder = "/Users/stoia1/Desktop/Uni/6_Semester/Forschungsmethoden/Forsch_meth/männlich/Stichprobe/TXT"
female_folder = "/Users/stoia1/Desktop/Uni/6_Semester/Forschungsmethoden/Forsch_meth/weiblich/TXT"

male_ttr_values = []
female_ttr_values = []


for filename in os.listdir(male_folder):
    with open(os.path.join(male_folder, filename), 'r', encoding='utf-8') as file:
        text = file.read()
        ttr = calculate_ttr(text)
        male_ttr_values.append(ttr)

for filename in os.listdir(female_folder):
    with open(os.path.join(female_folder, filename), 'r', encoding='utf-8') as file:
        text = file.read()
        ttr = calculate_ttr(text)
        female_ttr_values.append(ttr)

avg_ttr_male = sum(male_ttr_values) / len(male_ttr_values)
avg_ttr_female = sum(female_ttr_values) / len(female_ttr_values)

print("Durchschnittliches Typ-Token-Verhältnis (männlich):", avg_ttr_male)
print("Durchschnittliches Typ-Token-Verhältnis (weiblich):", avg_ttr_female)
```

***Was hat man herausgefunden:***

1) Durchschnittliche Satzlänge von Männer (durchschnitt_mann): 17.05 Wörter

2) Durchschnittliche Satzlänge von Frauen (durchschnitt_frau): 19.21 Wörter

3) Durchschn. Typ-Token-Verhältnis von Männer (avg_ttr_male): 0.17585858989142186

4) Durchschn. Typ-Token-Verhältnis von Frauen (avg_ttr_female): 0.15962798488389157

```python
import matplotlib.pyplot as plt

# Durchschnittliches TTR für Männer und Frauen
ttr_m = 0.17585858989142186
ttr_w = 0.15962798488389157

# Daten für das Tortendiagramm
labels = ['Männer', 'Frauen']
sizes = [ttr_m, ttr_w]
colors = ['blue', 'pink']

# Erstellen Sie ein Tortendiagramm
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Durchschnittliches Typ-Token-Verhältnis zwischen Frauen und Männern')

plt.show()
```
